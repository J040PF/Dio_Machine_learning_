```python
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dados para matriz
rótulos_reais = np.array([1, 0, 1, 2, 2, 0, 1, 2, 1])
rótulos_preditos = np.array([1, 0, 1, 2, 1, 0, 1, 2, 2])
```

```python
matriz_confusão = confusion_matrix(rótulos_reais, rótulos_preditos)
```

```python
sns.heatmap(matriz_confusão, annot=True, fmt='d', cmap='Blues',
            xticklabels=['','v','f'], yticklabels=['','v','f'])
plt.xlabel('Rótulos Preditos')
plt.ylabel('Rótulos Reais')
plt.title('Matriz de Confusão');
```

```python
# Atraves das formulas
vp = matriz_confusão[1,1]
vn = matriz_confusão[1,2]
fp = matriz_confusão[2,1]
fn = matriz_confusão[2,2]

acuracia = (vp+vn)/(vp+vn+fp+fn)
precisao = vp/(vp+fp)
sensibilidade = vp/(vp+fn)
especifidade = vn/(fp+vn)
fscore = 2*((precisao*sensibilidade)/(precisao+sensibilidade))

print(f'Acurácia: {acuracia:.2f}')
print(f'Precisão: {precisao:.2f}')
print(f'Sensibilidade: {sensibilidade:.2f}')
print(f'Especificidade: {especificidade:.2f}')
print(f'Fscore(F1): {fscore:.2f}')
```
